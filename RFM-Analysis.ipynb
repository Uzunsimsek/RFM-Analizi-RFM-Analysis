# RFM Skorları ile Müşteri Segmentasyonu

Recency, Frequency, Monetary ifadelerinin baş harflerinden oluşur.

Müşterilerin satın alma alışkanlıkları üzerinden pazarlama ve satış stratejileri belirlemeye yardımcı olan bir tekniktir.

Recency (yenilik): Müşterinin son satın almasından bugüne kadar geçen süre

- Diğer bir ifadesiyle “Müşterinin son temasından bugüne kadar geçen süre” dir.

- Bugünün tarihi - Son satın alma

- Örnek verecek olursak bugün bu analizi yapıyorsak bugünün tarihi - son ürün satın alma tarihi.

- Bu örneğin 20 olabilir 100 olabilir. Biliriz ki 20 olan müşteri daha sıcaktır. Daha son zamanlarda bizimle teması olmuştur.

Frequency (Sıklık): Toplam satın alma sayısı.

Monetary (Parasal Değer): Müşterinin yaptığı toplam harcama.

## RFM Analizi

```python
#Şimdi elde ettiğimiz verileri kısaca gözden geçirerek gerekli sınıflandırmaları ve puanlamaları yapacağız.
#shape metodu ile boyutları kontrol ediyoruz ve hepsinde eşit sayıda satır var
print(monetary_df.shape,recency_df.shape,freq_df.shape)
```

```python
#monetary,recency,freq dataframelerini concat fonkisyonu ile tek bir yerde birleştiriyoruz.
rfm=pd.concat([monetary_df,recency_df,freq_df], axis=1)
```

```python
rfm.head()
```

```python
#rfm RecencyScore adında bir değişken oluşturuyoruz.
#İçerisine Recency değişkenindeki gün sayılarına 1 ile 5 arasında puan veriyoruz.
#Burada yakın zamanda alışveriş yapılan değere 5, uzak zamanda yapılana 1 veriyoruz. 
#Örneğin: Recency alanında 324 yazan : son alışveriş 324 gün önce yaptı demek ve buna 1 veriyoruz. 
rfm["RecencyScore"] = pd.qcut(rfm["Recency"], 5, labels = [5, 4 , 3, 2, 1])
```

```python
rfm.head()
```

```python
#Buradada sıklık derecesini puanlıyoruz. Burada puanlamayı 1'den 5'e doğru yapıyoruz.
#Ne kadar sık alışveriş yapıyorsa iyidir.Çok sık alışveriş yapana 5, az yapana 1 puan şeklinde.
rfm["FrequencyScore"]= pd.qcut(rfm["Frequency"].rank(method="first"),5, labels=[1,2,3,4,5])
```

```python
rfm.head()
```

```python
rfm["MonetaryScore"]=pd.qcut(rfm["Monetary"],5,labels=[1,2,3,4,5])
```

```python
rfm.head()
```

```python
#müşteri numaralarına göre puanladığımız scoreları string değere çevirip yan yana topluyoruz
#burada rakamlar yan yana geldiğindeki değerleri gözlemliyoruz.
(rfm["RecencyScore"].astype(str)+rfm["FrequencyScore"].astype(str)+rfm["MonetaryScore"].astype(str)).head()
```

```python
#RfmScore adında bir değişken oluşturup yukarıda oluşturduğumuz puanlamayı gösteriyoruz.
rfm["RfmScore"]=(rfm["RecencyScore"].astype(str)+rfm["FrequencyScore"].astype(str)+rfm["MonetaryScore"].astype(str))
```

```python
rfm.head()
```

```python
#rfmscore değeri 555 olan değerleri gözlemliyoruz
#bunun bir nedeni yok gelip gelmiyor mu diye kontrol ettik:)
rfm[rfm["RfmScore"]=="555"].head()
```

```python
rfm.describe().T
```

```python
# Regular Expressions (Düzenli İfadeler) kullanılarak RFM haritası çıkarıldı
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
#RFM analizi yaparken tutarı dışarıda tutarak sıklık ve yenilik skorlarına göre işlem yapacağız.
#Bunun için sadece iki skoru görebilmek için Segment adlı bir değişken oluşturduk.
rfm["Segment"]= (rfm["RecencyScore"].astype(str)+rfm["FrequencyScore"].astype(str))
```

```python
rfm.head()
```

```python
#Yine rastgele bir değeri yazıp kontrol ediyoruz
rfm[rfm["Segment"]=="11"].head()
```

```python
#Yukarıda oluşturduğumuz rfm segment açıklamalarını segment numaralarının yerine geçiriyoruz
rfm["Segment"] = rfm["Segment"].replace(seg_map,regex=True)
```

```python
rfm.head()
```

```python
#segmentlere göre verilere göz atıyoruz
rfm[["Segment","Monetary","Recency","Frequency"]].groupby("Segment").agg(["mean","median","count"])
```

```python
rfm.shape[0]
```
